```python
# Tower of Hanoi

def hanoi(n, A, B, C):
    if n == 1:
        print("Move Disk 1 from", A, "to", C)
        return

    hanoi(n-1, A, C, B)
    print("Move Disk", n, "from", A, "to", C)
    hanoi(n-1, B, A, C)

n = int(input("Enter number of disks: "))

hanoi(n, 'A', 'B', 'C')
```
